```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#importing dataset
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
df.head()
```

```python
#GRE score out of 340 as X
X = df.iloc[:,1].values
X = np.reshape(X, (-1,1))
X
```

```python
#chances that grad student will get admitted on scale of 0 to 1
y = df.iloc[:,-1].values
y = np.reshape(y, (-1,1))
y
```

```python
lr = LinearRegression()
#training the regressor
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X,y)
```

```python
#predicting the chances if student has 250 in GRE
lr.predict([[250]])
```

```python
#plotting the actual dataset and regressor line that our model gave us
plt.plot(X,y,'o')
plt.plot(X,lr.predict(X))
plt.show()
```
